### les Listes

Les listes Python sont l'équivalent des tableaux  mais elle peuvent contenir des variables ayant des types  différents.

```python
liste = [3, True, "Hello",2.6]    # Création d'une liste
print(type(liste))
print(liste[2])  # les indices sont numérotés à partir de 0
print(liste[-1])     # on accede à la fin de la liste)
print("la liste contient",len(liste),"elements")    # on affiche le nombre d'éléments de la liste 
```

```python
liste.append(False) # append permet d'ajouter un élément à la fin de la liste
print(liste)  
liste.insert(0,30) # insert permet d'ajouter un élément à une position de la liste
print(liste)  
 
```

```python
x = liste.pop()     # pop enlève et retourne le dernier élément de la liste
print(x)
print(liste) 
```

```python
liste = liste.pop()      
print(liste)
```

```python
xs = [10,3,2.6] 
xs.sort() 
print(xs)
xs.extend([7,8,9])
print(xs)
 
```

```python
##### Slicing  ( acceder à des sous listes de la liste originale)
```

```python
tab=range(5)
print(tab)

liste = list(tab)   # list permet de créer une liste   (voir help(list)) 
print(liste)         
print(liste[1:3])    # afficher les éléments de la liste entre les indices 1 (inclus) et 3 (exclu)
print(liste[3:])     # dafficher tous les éléments de la liste à partir de l'indice 3
print(liste[:3])     # du début jusque l'indice 3 exclu (3 élements)
print(liste)         # toute la liste
print(liste[:])      # toute la liste

print(liste[:-1])    # du début jusque l'avant-dernier élément
print(len(liste))
print(liste[:len(liste)-1])    #  jusque l'avant-dernier élément

liste[2:4] = [8, 9]  # remplace une sous-liste  
print(liste)         
```

### Les Fonctions

Les fonctions Python sont définies par le mot-clé `def`

### Structure conditionnelle


```python
def comparaison(x,y):
    if x > y:
        return 'x est supérieur à y'
    elif x < y:
        return 'x est inférieur à y'
    else:
        return 'x et y sont égaux'

comparaison (0,3)   # appel de la fonction
```

```python
def parite(x):
    if x % 2==0:
        return 'x est paire'
    else:
        return 'x est impaire'
    
for x in [5, 10, 0,-2]:
    print(parite(x)) # appel de la fonction pour les 4 chiffres
```

## 4. Matplotlib

```python
Matplotlib est une bibliothèque graphique pour tracer des graphes
```

```python
import matplotlib.pyplot as plt
```

```python
X = np.array([0, 6])
Y = np.array([0, 250])

plt.plot(X, Y)
plt.show()

 
```

```python
X=np.array(range(10))
 
Y = np.power(X,2)
 

plt.plot(X, Y,marker="o")
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, 'o:r')     #marker|line|colorù
plt.grid(True)
plt.show()
```

```python
x = np.arange(0, 3 * np.pi, 0.1)
y = np.sin(x)

plt.figure()
plt.plot(x, y); 
```

```python
afficher plusieurs graphiques dans la même figure avec `subplot`:
```

```python

X = np.arange(0, 3 * np.pi, 0.1)
y1 =np.sin(X)
y2=X+2
print(y1) 
print(y2) 
plt.figure()
 
plt.subplot(2, 1, 1)

 
plt.plot(X, y1)
plt.title('y1')

 
plt.subplot(2, 1, 2)
plt.plot(X, y2)
plt.title('y2')

 
plt.show();
```

```python
size = 50
m = 1
b = 10
e = 4
x = 10*np.random.rand(size)
y = m*x + b + (2*e)*np.random.rand(size)-e
plt.grid(True)

plt.scatter(x, y)
plt.ylabel(r'$y$', fontsize = 60)
plt.xlabel(r'$x$', fontsize = 20)
plt.legend(['scatter plot'])
```

```python
from mpl_toolkits import mplot3d
fig = plt.figure()
ax = plt.axes(projection='3d')
```

```python
from mpl_toolkits import mplot3d
ax = plt.axes(projection='3d')

# Data for a three-dimensional line
zline = np.linspace(0, 15, 1000)
xline = np.sin(zline)
yline = np.cos(zline)
ax.plot3D(xline, yline, zline, 'gray')

# Data for three-dimensional scattered points
zdata = 15 * np.random.random(100)
xdata = np.sin(zdata) + 0.1 * np.random.randn(100)
ydata = np.cos(zdata) + 0.1 * np.random.randn(100)
ax.scatter3D(xdata, ydata, zdata, c=zdata, cmap='Greens');
```
