##  avec skilearn


```python
from sklearn.linear_model import Perceptron 
from sklearn import metrics 
from sklearn.metrics import confusion_matrix 

```

```python
X_train=np.array([[ -4, 1], [4, 2],[ 3,-2],[ -4, -2],[ -1, -1.5],[ 4, -4]])
y_train=np.array([1, 1,1,0,0,0])
print(X_train)
print(y_train)
```

```python
# Create a Perceptron model 
perceptron = Perceptron(max_iter=20, eta0=0.1)

# Train the model 
perceptron.fit(X_train, y_train)

# Make predictions 
y_pred = perceptron.predict(X_train) 
print(y_pred)
```

```python
# afficher les poids du model
print('model weights:')
print(perceptron.coef_)
print()
print('bias / intercept weights')
print(perceptron.intercept_)
```

```python
# Calculate confusion matrix 
conf_matrix = confusion_matrix(y_train, y_pred) 
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = conf_matrix, display_labels = [False, True]) 
  
cm_display.plot() 
plt.show()
```

## avec sklearn

```python
import numpy as np
import sklearn.neural_network

inputs = np.array([[0,0],[0,1],[1,0],[1,1]])
expected_output = np.array([0,1,1,0])

model = sklearn.neural_network.MLPClassifier(
                activation='logistic',
                max_iter=100,
                hidden_layer_sizes=(5,),
                solver="lbfgs")  #lbfgs
model.fit(inputs, expected_output)
print('predictions:', model.predict(inputs))
```

```python
# afficher les poids du model
print('model weights:')
print(model.coefs_)
 
```
